```python
import numpy as np
import matplotlib.pyplot as plt
from impedance import preprocessing 
from impedance.visualization import plot_nyquist
from impedance.visualization import plot_bode
from impedance.visualization import plot_residuals
from impedance.models.circuits import CustomCircuit
```

```python
import pandas as pd
import os 
```

```python
data = pd.read_csv('12.02.21_D', skiprows=9, sep='\t', header=None, 
                   names=['Temp', 'NA', 'NA2', 'Cap', 'Loss', 'Freq']
                  )
data.head()
```

```python
maxT = data['Temp'].max()
idxStart = data[(data['Temp'] > maxT - 1)].index[0]
idxStart
cold = data.loc[idxStart:,:] #selects data from highest temp to end and all columns
#Junk Cutoffs-- Cap between 0 and 1, Freq >1. May need adjustment depending on data
HiCap = cold[cold['Cap']<1]
LowCap = HiCap[HiCap['Cap']>0]
fd = LowCap[LowCap['Freq']>1] #filtered data
fd.head()
```

```python
#plots for 200C
tdf = fd[(fd['Temp']>199)&(fd['Temp']<201)]
th = 0.0005
Diameter = 0.00521
AEl = np.pi*(Diameter/2)**2
perm = 8.85*10**(-12)
Z200 = (1/(1j*tdf['Freq'].values*2*np.pi*perm*AEl/th*(tdf['Cap'].values*th/(AEl*perm)-(1j*tdf['Loss'].values*tdf['Cap'].values*th)/(AEl*perm))))/1e5
RZ200 = np.real((1/(1j*tdf['Freq'].values*2*np.pi*perm*AEl/th*(tdf['Cap'].values*th/(AEl*perm)-(1j*tdf['Loss'].values*tdf['Cap'].values*th)/(AEl*perm))))/1e5)
IZ200 =np.imag((1/(1j*tdf['Freq'].values*2*np.pi*perm*AEl/th*(tdf['Cap'].values*th/(AEl*perm)-(1j*tdf['Loss'].values*tdf['Cap'].values*th)/(AEl*perm))))/1e5)
```

```python
fig,ax = plt.subplots(1,2,figsize = (8,3))
plot_bode(ax,tdf['Freq'].values,Z200)
fig,ax = plt.subplots()
plot_nyquist(ax,Z200)
```

```python
#t = R*C, f = 1/2pi*t = 1/2piRC, so C = 1/2piRf
circuit = CustomCircuit('p(R1,CPE1)-p(R2,C2)',initial_guess = [65000/1e5,1e5/(2*np.pi*65000),0.84,31000/1e5,1e5/(2*np.pi*31000)])
circuit.fit(tdf['Freq'].values,Z200)
```

```python
circuit.parameters_
print(circuit)#circuit is an object, so we want to figure out what the object model is (something like that?)
Z200_fit = circuit.predict(tdf['Freq'].values)
```

```python
# makes new dataframe 
dataFit = pd.DataFrame(columns = ['Temp','R1','CPE1_0','CPE1_0','R2','C2'])
dataFit.head()
```

```python
# create array to insert into new dataframe from above. using numpy.insert because circuit.params is a numpy.array
fit_params = np.insert(circuit.parameters_, 0, [200])
print(fit_params)
```

```python
# insert data into dataframe. Need to do some more googling to figure this out. also put some time on the calendar to work onthis with Tyler!
dataFit.append(pd.DataFrame(fit_params),ignore_index = True)
# dataFit.head()
```

```python
fig, ax = plt.subplots()
plot_nyquist(ax, Z200, fmt='o')
plot_nyquist(ax, Z200_fit, fmt='-')
plt.legend(['Data','Fit'])
plt.show()
```
